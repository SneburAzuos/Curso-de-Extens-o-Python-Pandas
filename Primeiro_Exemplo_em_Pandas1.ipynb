```python
#Importando biblioteca pandas
import pandas as pd
```

```python
#quais são as coluas e os tipos de dados
display(combustiveis_df.info())
```

```python
#mostras informações estatísticas básicas
display(combustiveis_df.describe())
```

```python
#Como utilizar uma unica coluna
display(combustiveis_df['Revenda'])
```

```python
#Criando um novo dataframe com as colunas que eu quero
ca_df = combustiveis_df[['Revenda' , 'Municipio' , 'Produto' , 'Valor de Venda']]
display(ca_df)
```

```python
#exibe a 4a. linha
display(ca_df.loc[3])
```

```python
#Décima até a vigésima linha
display(ca_df.loc[9:19])
```

```python
#Criar um dataframe  gas_df 
#contendo apenas 4 colunas (Revenda, Municipio, Produto Valor de Venda)
#somente gasolina
#exibir na tela
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#Exibir Posto e municipio da gasolina mais cara do pais
display(gas_df[['Revenda' , 'Municipio' , 'Valor de Venda' , ]].max())
```

```python
#Localizar linha com multiplas condições para filtragem
#Quais são os preços, postos que vendem ETANOL na minha cidade INDAIATUBA.
 #Ordenar do menor para o maior
etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
display (etanol_indaiatuba_df.sort_values(by = 'Valor de Venda' ))
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO?
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), 
                            ['Valor de Venda']].mean())
```

```python
#Segunda Opção: Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO?
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
#Como mostrar média do valor de venda por COMBUSTÍVEL Brasil todo?
media_por_combustivel_df = ca_df[['Produto' , 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel_df)
```

```python
#Quero adicionar uma columa booleana no combustiveis_df chamada "Ativo"
#que, por padrão, vai ser True para TODAS as linhas
combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())
```

```python
# Coluna nova chamda "Obs"
# que preencha automaticamente "MELHOR CIDADE"
# quando O Municipio for Indaiatuba, SP...
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') & ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA'))])
```

```python
#exportar dataframe etanol_indaiatuba para o excel
etanol_indaiatuba_df.to_excel("etanol_indaiatuba.xlsx", sheet_name='Etanol em Indaiatuba')
```
