```python
#Criando um novo dataframe com as colunas que eu quero
ca_df = combustiveis_df[['Revenda' , 'Municipio' , 'Produto' , 'Valor de Venda']]
display(ca_df)
```

```python
#Criar um dataframe  gas_df 
#contendo apenas 4 colunas (Revenda, Municipio, Produto Valor de Venda)
#somente gasolina
#exibir na tela
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#Exibir Posto e municipio da gasolina mais cara do pais
display(gas_df[['Revenda' , 'Municipio' , 'Valor de Venda' , ]].max())
```

```python
#Localizar linha com multiplas condições
vendas_gasolina_df = combustiveis_df.loc[(combustiveis_df['Produto'] == 'Gasolina') & (combustiveis_df['Municipio'] == 'Indaiatuba')]
display(vendas_gasolina_df)
```

```python
#Loc condicional especificando colunas
vendas_gasolina_df = combustiveis_df.loc[(combustiveis_df['Produto'] == 'Gasolina') & (combustiveis_df['Municipio'] == 'Indaiatuba'), 
                                         ['Municipio' , 'Revenda' , 'Data da Coleta' , 'Produto' , 'Valor de Venda']]
display(vendas_gasolina_df)
```

```python
#Informação nova a partir de uma coluna que existe
vendas_gasolina_df['Imposto'] = vendas_gasolina_df['Valor de Venda'] * 0.10
display(vendas_gasolina_df)
```

```python
#removendo colunas - axis=1 é coluna/axis=0 é linha
vendas_gasolina_df = vendas_gasolina_df.drop("Imposto" , axis=1)
display(vendas_gasolina_df)
```

```python
#deletar linhas e colunas completamente vazias
vendas_gasolina_df = vendas_gasolina_df.dropna(how='all')
display(vendas_gasolina_df)
```

```python
#deletar colunas completamente vazias
vendas_gasolina_df = vendas_gasolina_df.dropna(how='all' , axis=1)
display(vendas_gasolina_df)
```

```python
copiar caminho tb funcionaendas_gasolina_df['Imposto'] = vendas_gasolina_df['Valor de Venda'] * 0.10

#preencherautomático com a média da coluna
vendas_gasolina_df['Imposto'] = vendas_gasolina_df['Imposto'].fillina(vendas_gasolina_df['Imposto'].mean())
display(vendas_gasolina_df)
```

```python
#média por combutível
media_por_produto = combustiveis_df[['Produto' , 'Valor de Venda']].groupby('Produto').mean()
display(media_por_produto)
```
